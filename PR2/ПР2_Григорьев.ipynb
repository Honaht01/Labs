```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://www.nasdaq.com/market-activity/index/comp/historical?page=1&rows_per_page=10&timeline=m1'
response = requests.get(url)
page_content = response.content
```

Доступ к сайту NASDAQ получить не возможно, блок выполняется бесконечно. Было принято решение заменить сайт.

```python
url = 'https://www.investing.com/indices/major-indices'
response = requests.get(url)
page_content = response.content
```

```python
soup = BeautifulSoup(page_content, 'html.parser')
table = soup.find('tbody', {'class': 'datatable-v2_body__8TXQk'})
rows = table.find_all('tr')
```

```python
data = []
for row in rows[1:]: # Пропускаем заголовок
 cols = row.find_all('td')
 cols = [ele.text.strip() for ele in cols]
 data.append(cols)
df = pd.DataFrame(data, columns=['#', 'Name', 'Last', 'High', 'Low', 'Change', 'Change%', 'Time'])
df['Last'] = df['Last'].str.replace(',', '').astype(float)
df['High'] = df['High'].str.replace(',', '').astype(float)
df['Low'] = df['Low'].str.replace(',', '').astype(float)
```

```python
df.head(50)
```

```python
df.info()
```

```python
df.plot(kind='bar', x='Name', y='Last', legend=False)
plt.title('Current indexes')
plt.xlabel('Index')
plt.ylabel('Current value')
plt.show()
```

```python
top_10 = df.nlargest(10, 'Last')
plt.figure(figsize=(8, 8))
plt.pie(top_10['Last'], labels=top_10['Name'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 last values by indexes')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
top_10 = df.nlargest(10, 'High')
plt.figure(figsize=(8, 8))
plt.pie(top_10['High'], labels=top_10['Name'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 highest values by indexes')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Выводы:
1. Рынок достаточно стабилен, самое большое изменение индекса - увеличение на 2.14%
2. Самые большие на данный момент индексы близки к своим наибольшим историческим значениям (в рамках зафиксированных на сайте)
3. Наибольшие индексы, такие как Bovespa, BSE Sebsex, Karachi 100 и другие, сохраняют свои позиции.
